```python
def parse_NYT_query(data):
    '''
    Takes an inputted JSON-loaded string of NYT API results and parses it into a DataFrame
    
    Input: JSON-loaded NYT response string
    Output: Pandas DataFrame of response string
    '''
    
    results = []
    for article in data.get('response').get('docs'):
        result = {'id': article.get('_id'),
                  'source': article.get('source'),
                  'news_desk': article.get('news_desk'),
                  'blog': article.get('blog'),
                  'pub_date': article.get('pub_date'),
                  'url': article.get('web_url'),
                  'abstract': article.get('abstract'),
                  'lead_paragraph': article.get('lead_paragraph'),
                  'snippet': article.get('snippet'),
                  'word_count': article.get('word_count'),
                  'keywords': article.get('keywords')}
        
        # try parsing headline, otherwise simply pass entire element
        try:
            result['headline'] = article.get('headline').get('main')
            result['seo_headline'] = article.get('headline').get('seo')
        except:
            result['headline'] = article.get('headline')
            result['seo_headline'] = article.get('headline')
        
        # try parsing author, otherwise simply pass entire element
        try:
            result['author'] = article.get('byline').get('original')[3:]
        except:
            result['author'] = article.get('byline')
        
        # append the results to growing list
        results.append(result)
    
    return pd.DataFrame(results)
    
def NYT_article_search(options = {'fq': 'subject:("Presidential Election of 2012") AND type_of_material:("News")',
                                  'end_date': '20121106'},
                       api_key = 'b5b3713fb7bf65301ab0efb581a328b4:15:61099097',
                       full_query = False,
                       start_page = 1,
                       print_result = False):
    '''
    Retrieves NYT articles from API based on keywords and returns a DataFrame of articles
    
    Input: dictionary of NYT parameters (http://developer.nytimes.com/docs/read/article_search_api_v2)
        q: search word
        fq: field query. Example: 'fq': 'subject:("Presidential Election of 2012") AND type_of_material:("News")'
        begin_date: YYYYMMDD
        end_date: YYYYMMDD
        ...etc.
    Output: Pandas DataFrame of all search results
    '''
    
    # run the query once to determine how many pages to call
    url = 'http://api.nytimes.com/svc/search/v2/articlesearch.json'
    options['page'] = start_page
    options['api-key'] = api_key
    data = json.loads(requests.get(url, params=options).text)
    
    # print the data if asked for it
    if(print_result): pprint.pprint(data)
    
    # calculate the page limit -- max allowed currently by NYT is 100
    page_limit = data['response']['meta']['hits'] / data['response']['meta']['offset'] + 1
    if (page_limit > 100):
        print "NYT allows for up to 100 pages to be queried; request of %i will be truncated to 100." % page_limit
        page_limit = 100
    else:
        print "Querying %i pages..." % page_limit
    
    # parse the first response
    NYT_df = parse_NYT_query(data)
    
    # if full query desired, run the same query page_limit times, parsing into a DataFrame and concatenating the results
    if(full_query):
        for iteration in range(2, page_limit + 1, 1):
            
            if(iteration % 25 == 0): 
                print "Pages complete: %i / %i" % (iteration, page_limit)
            
            # append current page to options dictionary
            options['page'] = iteration
            options['api-key'] = api_key
            data = requests.get(url, params=options).text
            data = json.loads(data)
            
            # attempt to parse response, if successful, add to DataFrame. If not, print it.
            try:
                NYT_df = NYT_df.append(parse_NYT_query(data), ignore_index=True)
            except:
                pprint.pprint(data)
    
    return NYT_df
```

###USA Today API

```python
# USA Today (1000 calls a day)

def parse_USA_Today(data):
    '''
    Takes an inputted JSON-loaded string of USA Today API results and parses it into a DataFrame
    
    Input: JSON-loaded USA Today response string
    Output: Pandas DataFrame of response string
    '''
    
    results = []
    try:
        for article in data.get('stories'):
            result = {'headline': article.get('title'),
                      'summary': article.get('description'),
                      'url': article.get('link'),
                      'pubdate': datetime.datetime.strptime(article.get('pubDate'), '%a, %d %b %Y %H:%M:%S %Z'),
                      'source': 'USA Today',
                      'metadata': article.get('guid')}
        
            # append the results to growing list
            results.append(result)
    except:
        print "Failed to parse: "
        pprint.pprint(data)
    
    return pd.DataFrame(results)

def USA_Today_article_search(options = {'tag': 'elections-2012',
                                      'api_key': 'm5bqprbn2hjd9ayfev4sy8m4',
                                      # 'wz2zcterwj8e3y68wacyhwwk', # alternate 'djp3yvuhr5zsun3m9yfybcjx',
                                      'encoding': 'json',
                                      'page': 0,
                                      'fromdate': '2011-01-01',
                                      'todate': '2012-11-30'},
                           print_result = False,
                           full_query = False):
    '''
    Retrieves USA Today articles from API based on keywords and returns a DataFrame of articles
    
    Input: dictionary of USA Today parameters (http://developer.usatoday.com/docs/read/articles)
        search: keyword search
        tag: curated keyword search
        encoding: json
        fromdate / todate: YYYY-MM-DD
    Output: Pandas DataFrame of all search results
    '''
    
    # run the query up to 1000 times or until the query returns the same data multiple times
    old_data, new_data = '', ''
    USA_Today_full = pd.DataFrame()
    current_page, errors = int(options['page']), 0
    while (current_page < 3):
        # run the query
        url = 'http://api.usatoday.com/open/articles'
        options['page'] = str(current_page)

        # try loading a given page -- if it fails, try again. Otherwise go to the next page.
        try:
            new_data = json.loads(requests.get(url, params=options).text)
        except:
            try:
                new_data = json.loads(requests.get(url, params=options).text)
            except:
                print "Page %i failed to load." % current_page
                current_page += 1
                errors += 1
                
                # if there have been more than 10 errors in a row, break. Otherwise continue
                if(errors > 10): 
                    break 
                else: 
                    continue
        
        if(current_page % 25 == 0): 
                print "Pages complete: %i" % current_page
        
        # parse data into DataFrame (may require multiple pages as in NYT)
        USA_Today_single = parse_USA_Today(new_data)
        
        # compare new data to old data, if the same, break. Else, add to DF and reset errors to zero
        if(new_data == old_data):
            print "Two pages returned the same data and therefore the query has likely ended."
            break
            
        else:
            if(USA_Today_full.shape[0] > 0):
                USA_Today_full = USA_Today_full.append(USA_Today_single, ignore_index = True)
            else:
                USA_Today_full = USA_Today_single
            
            # save new data as old data for future comparison and increment current_page
            old_data = new_data
            current_page += 1
            errors = 0
            
            # if full_query is false, only run query once
            if(full_query == False):
                break
    
    return USA_Today_full
```

###Wall Street Journal Scraping

The WSJ does not have an API: however, it does have a nice easily-accessible archive of the form "http://online.wsj.com/public/page/archive-YYYY-M(M)-D(D).html", and within each of these pages there is a list of articles.

```python
def WSJ_single_page(date):
    '''
    Scrapes a single page of the WSJ article archive.
    
    Inputs: YYYY-M(M)-D(D) of desired archive
    Output: Pandas DataFrame of results
    '''
    
    # generate URL and load the page -- if it fails, return None
    url = 'http://online.wsj.com/public/page/archive-%s.html' % date
    
    try:
        html = requests.get(url).text
    except:
        print "The site %s did not load." % url
        return None
    
    # declare dom object to begin parsing the data and initialize DataFrame
    dom = web.Element(html)
    
    WSJ_result_df = pd.DataFrame()
    
    # iterate through all <li>'s for articles, generating dicts
    return pd.DataFrame([{'source': 'Wall Street Journal',
                          'headline': article('h2 a')[0].content,
                          'url': article('h2 a')[0].attributes.get('href'),
                          'snippet': article('p')[0].content,
                          'pub_date': pd.datetime(int(date.split('-')[0]), int(date.split('-')[1]), int(date.split('-')[2]))} 
                          for article in dom('ul.newsItem li')])

def WSJ_scrape_archives(start_date, end_date = datetime.datetime.now()):
    '''
    Scrape all articles, headlines, and descriptions from the Wall Street Journal Archives
    
    Inputs: begin_date, end_date in YYYY-M(M)-D(D) format (use datetime.datetime(year, month, day))
    Output: DataFrame with all inputs appended
    '''
    
    delta = end_date - start_date
    
    WSJ_full = pd.DataFrame()
    
    # for every date in the given range, add the new results to a DataFrame
    for i in range(delta.days + 1):
        
        # print update every 30 days (roughly a month)
        if (i % 30 == 0):
            print "Currently scraping: " + (start_date + td(days=i)).strftime("%Y-%-m-%-d")
        
        # query and parse the page
        WSJ_single = WSJ_single_page((start_date + td(days=i)).strftime("%Y-%-m-%-d"))
        
        # if full is empty, set equal. Otherwise append.
        if(WSJ_single is not None):
            if(WSJ_full.shape[0] > 0):
                WSJ_full = WSJ_full.append(WSJ_single, ignore_index = True)
            else:
                WSJ_full = WSJ_single
    
    return WSJ_full
```

```python
# Bing Search API from https://github.com/xthepoet/pyBingSearchAPI
class BingSearchAPI():
    bing_api = "https://api.datamarket.azure.com/Data.ashx/Bing/Search/v1/Composite?"
    
    def __init__(self, key):
        self.key = key

    def replace_symbols(self, request):
        # Custom urlencoder.
        # They specifically want %27 as the quotation which is a single quote '
        # We're going to map both ' and " to %27 to make it more python-esque
        request = string.replace(request, "'", '%27')
        request = string.replace(request, '"', '%27')
        request = string.replace(request, '+', '%2b')
        request = string.replace(request, ' ', '%20')
        request = string.replace(request, ':', '%3a')
        return request
        
    def search(self, sources, query, params):
        ''' This function expects a dictionary of query parameters and values.
            Sources and Query are mandatory fields. 
            Sources is required to be the first parameter.
            Both Sources and Query requires single quotes surrounding it.
            All parameters are case sensitive. Go figure.

            For the Bing Search API schema, go to http://www.bing.com/developers/
            Click on Bing Search API. Then download the Bing API Schema Guide
            (which is oddly a word document file...pretty lame for a web api doc)
        '''
        request =  'Sources="' + sources    + '"'
        request += '&Query="'  + str(query) + '"'
        for key,value in params.iteritems():
            request += '&' + key + '=' + str(value) 
        request = self.bing_api + self.replace_symbols(request)
        return requests.get(request, auth=(self.key, self.key))
```

```python
def NYDN_single_page(date):
    '''
    Scrapes a single page of the NY Daily News article archive, finding URLs of Politics pages
    and placing those pages into a DataFrame
    
    Inputs: YYYYMMDD of desired archive
    Output: Pandas DataFrame of results
    '''
    
    # generate URL and load the page -- if it fails, return None
    url = 'http://www.nydailynews.com/archives?pdate=%s' % date
    
    try:
        html = requests.get(url).text
    except:
        print "The site %s did not load." % url
        return None
    
    # declare dom object to begin parsing the data and initialize DataFrame
    dom = web.Element(html)
    
    # iterate through all dl's on page, storing the urls that fall under Politics
    politics_section = False
    url_list = []
    for dl in dom('dl'):
        for row in dl('*'):
            # start collecting URLs at the politics section and end at Crime or Football
            if(row.content == 'Politics'):
                politics_section = True
            if(row.content == 'Crime' or row.content == 'Football'):
                politics_section = False
            
            # add the URL to the list
            if(politics_section and len(row('a')) > 0):
                url_list.append(row('a')[0].attributes['href'])
    
    # for each of the URLs in the list, download the page and parse data from it
    NYDN_results = []
    for sub_url in url_list:    
        # attempt to download the given page
        try:
            html = requests.get(sub_url).text
        except:
            print "The site %s did not load." % sub_url
            continue
        
        # declare a dom element and parse data from page
        dom = web.Element(html)
        try:
            NYDN_result = {'headline': dom('h1.story-header')[0].content,
                           'byline': plaintext(dom('h3.byline')[0].content),
                           'summary': dom('h2.story-subheader')[0].content,
                           'pub_date': pd.datetime(int(date[:4]), int(date[4:6]), int(date[6:8])),
                           'body': plaintext(dom('div#story')[0].content),
                           'url': sub_url,
                           'source': 'New York Daily News'}
        except:
            print "Parsing page %s failed." % sub_url
            continue
        
        # add to list of single page dicts
        NYDN_results.append(NYDN_result)
    
    return pd.DataFrame(NYDN_results)

def NYDN_scrape_archives(start_date, end_date = datetime.datetime.now()):
    '''
    Scrape all articles, headlines, and descriptions from the New York Daily News Archives
    
    Inputs: begin_date, end_date in YYYYMMDD format (use datetime.datetime(year, month, day))
    Output: DataFrame with all inputs appended
    '''
    
    delta = end_date - start_date
    
    NYDN_full = pd.DataFrame()
    
    # for every date in the given range, add the new results to a DataFrame
    for i in range(delta.days + 1):
        
        # print update every 30 days (roughly a month)
        if (i % 30 == 0):
            # store a copy of the file
            NYDN_full.to_csv("./NYDN_data/NYDN_buffer.csv", encoding = "UTF-8")
            print "Currently scraping: " + (start_date + td(days=i)).strftime("%Y%m%d")
        
        # query and parse the page
        NYDN_single = NYDN_single_page((start_date + td(days=i)).strftime("%Y%m%d"))
        
        # if full is empty, set equal. Otherwise append.
        if(NYDN_single is not None):
            if(NYDN_full.shape[0] > 0):
                NYDN_full = NYDN_full.append(NYDN_single, ignore_index = True)
            else:
                NYDN_full = NYDN_single
    
    return NYDN_full
```

###PQ Archiver Scraping
The Boston Globe, the Washington Post, the Wall Street Journal (redux), USA Today, the Los Angeles Times, etc. all use the ProQuest Archiver system (http://www.pqarchiver.com/index.html) to store articles. We wrote a similar scraper for each of these newspapers, one example is below.

```python
def PQarchive_url_list(start_date, end_date, page, newspaper_tag = 'latimes', query = 'romney OR obama', debug = False):
    '''
    Scrapes the PQ archive system to get a list of all URLs.
    
    Inputs: M(M)-D(D)-YYYY of start and end date, page number (1-indexed)
    Output: URL List
    '''

    # split dates into M, D, Y
    start_date = start_date.split('-')
    end_date = end_date.split('-')
    
    options = {}
    
    # run the query
    url = 'http://pqasb.pqarchiver.com/' + newspaper_tag + '/results.html'
    options['st'] = 'advanced'
    options['sortby'] = 'CHRON'
    options['datetype'] = 6
    options['frommonth'] = start_date[0]
    options['fromday'] = start_date[1]
    options['fromyear'] = start_date[2]
    options['tomonth'] = end_date[0]
    options['today'] = end_date[1]
    options['toyear'] = end_date[2]
    options['type'] = 'current'
    options['start'] = (page-1)*10
    options['QryTxt'] = query
    
    # try to get url with specified parameters
    try:
        r = requests.get(url, params=options)
        html = r.text
        if debug: print r.url
    except:
        print 'Unable to parse URL list for ' + str(url)
        return None
    
    # declare dom object to begin parsing the data
    dom = web.Element(html)
    
    url_list = []
    
    # find each url
    for a in dom('.result_title a'):
        url_list.append(str(a.attrs['href']))

    return url_list

def PQarchive_scrape_archives(start_date, end_date, newspaper_tag = 'latimes', source = 'LA Times', 
                              query = 'romney OR obama', debug = False):

    '''
    Scrape all articles, headlines, and descriptions from the PQarchive system
    
    Inputs: begin_date, end_date in YYYY-M(M)-D(D) format (use datetime.datetime(year, month, day))
    Output: DataFrame with all inputs appended
    '''
    
    url_list = []
    
    # start page counter at 1
    i = 1
    
    # scrape first page of results
    new_url_list = PQarchive_url_list(start_date, end_date, i, newspaper_tag,  query, debug)
    
    # keep scraping until we run out of pages
    while new_url_list:
    
        # add 10 new urls to url_list
        url_list += new_url_list
        
        # Move to next page and scrape it
        i += 1
        new_url_list = PQarchive_url_list(start_date, end_date, i, newspaper_tag, query, debug)
    
    # create DataFrame
    PQarchive_full = pd.DataFrame(columns=['author','date','section','word_count','abstract','headline','source', 'subjects'])
    
    url_count = 0
    
    print "This query will run for "  + str(len(list(set(url_list))))
    
    if debug: print list(set(url_list))
    
    # loop through complete url list
    for url in list(set(url_list)):

        url_count += 1
        if url_count % 100 == 0:
            print url_count
        
        # get html from each url
        try:
            html = requests.get('http://pqasb.pqarchiver.com' + url).text
        except:
            print "Could not retrieve html for page " + str(url)
            continue
    
        # declare dom object to begin parsing the data
        dom = web.Element(html)

        prev_content = ''
        article = {}
        
        # don't scrape if no abstract
        if dom('p'):
            
            article['abstract'] = ''
            # add the abstract text to the article dict
            for p in dom('p'):
                if (p.content != '<a href="http://www.latimes.com/terms">Terms of Service</a> | <a href="http://www.latimes.com/privacypolicy">Privacy Policy</a> | Los Angeles Times, 202 West 1st Street, Los Angeles, California, 90012 | Copyright 2011'):
                    article['abstract'] += ' ' + plaintext(p.content)
        
            # go to each table row
            for td in dom('td'):
           
                # since we're going in order of td's, 'Author:' will come right before the author's name, and so on
                if prev_content == 'Author:':
                    article['author'] = td.content
                elif prev_content == 'Date:':
                    article['date'] = td.content
                elif prev_content == 'Section:':
                    article['section'] = td.content
                elif prev_content == 'Text Word Count:':
                    article['word_count'] = td.content
                elif prev_content == 'Subjects:':
                    article['subjects'] = plaintext(td.content)
                
                # update prev_content for the next loop
                prev_content = td.content
            
            # append each article dict as a row to the full df, adding headline and source
            for headline in dom('.docTitle'):
                article['headline'] = headline.content
            if debug: print urlparse.parse_qs(urlparse.urlparse(url).query)
            article['source'] = urlparse.parse_qs(urlparse.urlparse(url).query)['pub'][0]
            PQarchive_full = PQarchive_full.append(article, ignore_index = True)
        
    return PQarchive_full
```

###Newspaper Data Cleaning
Now that we have gathered thousands of articles from many different news sources, it is necessary to combine, clean, and sort that data. We then split the full dataset into two groups: articles that likely mostly concern Romney, and articles that likely mostly concern Obama.

```python
def globe_date_convert(date):
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    temp = date.split(' ')
    
    # convert month name to number
    month_num = 0
    
    month_count = 0
    for month in months:
        month_count = month_count + 1
        if temp[0] == month:
            month_num = month_count
            
    #get the day
    day_num = temp[1].replace(',','')
    date_result = datetime.date(int(temp[2]), int(month_num), int(day_num)) 
    return date_result

def guardian_date_convert(date):
    temp = date.split('T')
    date_result = datetime.datetime.strptime(temp[0], "%Y-%m-%d").date()
    return date_result

def nydn_date_convert(date):
    temp = date.split(' ')
    date_result = datetime.datetime.strptime(temp[0], "%Y-%m-%d").date()
    return date_result


# clean Chicago Tribune
chicago_frame = pd.read_csv("./Data/Newspaper_data/Chicago_Tribune_Data/ChicagoTribune_romney_or_obama_newer.csv", 
                            encoding = "UTF-8")
chicago_frame['date'] = chicago_frame.date.apply(globe_date_convert)

# Chcicago Tribune did not have "search for Obama or Romney" option, therefore this will be performed manually
filter_pattern = "Obama|Romney"
chicago_frame = chicago_frame[np.logical_or(chicago_frame['abstract'].str.contains(filter_pattern), 
                                            chicago_frame['headline'].str.contains(filter_pattern))]

# clean globe data
globe_frame = pd.read_csv("./Data/Newspaper_data/Globe_Data/Globe_romney_or_obama.csv", encoding = "UTF-8")
globe_frame['date'] = globe_frame.date.apply(globe_date_convert)

# clean guardian data
guardian_frame = pd.read_csv("./Data/Newspaper_data/Guardian_Data/Guardian_2012_election_full.csv", encoding = "UTF-8")
guardian_frame = guardian_frame.drop(['id','trailText','url'],1)

guardian_frame['date'] = guardian_frame.pub_date.apply(guardian_date_convert)
guardian_frame = guardian_frame.drop('pub_date',1)
guardian_frame.rename(columns={'standfirst': 'abstract', 'sectionName':'section'}, inplace=True)
guardian_frame['source'] = 'Guardian'

# clean LATimes data
latimes_frame = pd.read_csv("./Data/Newspaper_data/LATimes_Data/LATimes_romney_or_obama_total.csv", encoding = "UTF-8")
latimes_frame['date'] = latimes_frame.date.apply(globe_date_convert)

# cleaning Newsday
newsday_frame = pd.read_csv("./Data/Newspaper_data/Newsday_Data/newsday_romney_or_obama_newer.csv", encoding = "UTF-8")
newsday_frame['date'] = newsday_frame.date.apply(globe_date_convert)

# clean nydn data
nydn_frame = pd.read_csv("./Data/Newspaper_data/NYDN_Data/NYDN_politics_2008_present.csv", encoding = "UTF-8")
nydn_frame = nydn_frame.drop(['body', 'url'],1)
nydn_frame.rename(columns={'summary':'abstract','pub_date':'date', 'byline':'author'}, inplace=True)

nydn_frame['date'] = nydn_frame.date.apply(nydn_date_convert)
nydn_frame = nydn_frame[nydn_frame['source'] == 'New York Daily News']

nydn_frame = nydn_frame[nydn_frame['date'] > datetime.date(2012,1,1)]
nydn_frame = nydn_frame[nydn_frame['date'] < datetime.date(2012,11,6)]

# NYDN did not have "search for Obama or Romney" option, therefore this will be performed manually
filter_pattern = "Obama|Romney|obama|romney"
nydn_frame = nydn_frame[np.logical_or(nydn_frame['abstract'].str.contains(filter_pattern), 
                                      nydn_frame['headline'].str.contains(filter_pattern))]

# clean NYT data
nyt_frame = pd.read_csv("./Data/Newspaper_data/NYT_Data/NYT_2012_election_full.csv", encoding = "UTF-8")
nyt_frame['date'] = nyt_frame.pub_date.apply(guardian_date_convert)

# if no abstract, add snippet else add abstract
abstract_list = []
for row in nyt_frame.iterrows():
    if(pd.isnull(row[1][1]) or row[1][1] == ''):
        abstract_list.append(row[1][11])
    else:
        abstract_list.append(row[1][1])

nyt_frame['abstract'] = abstract_list
nyt_frame = nyt_frame.drop(['blog','pub_date','id', 'url','lead_paragraph','seo_headline','snippet'],1)
nyt_frame.rename(columns={'news_desk':'section'},inplace=True)

# cleaning PANewspapers
pa_frame = pd.read_csv("./Data/Newspaper_data/PA_newspapers_Data/PA_romney_or_obama_newer.csv", encoding = "UTF-8")
pa_frame['date'] = pa_frame.date.apply(globe_date_convert)

# clean Washington Post data
wp_frame = pd.read_csv("./Data/Newspaper_data/WashingtonPost_Data/WP_election_2012.csv", encoding = "UTF-8")
wp_frame['date'] = wp_frame.date.apply(globe_date_convert)

# clean USAToday
usa_t_frame = pd.read_csv("./Data/Newspaper_data/USA_Today_data/USAToday_romney_or_obama.csv", encoding = "UTF-8")
usa_t_frame['date'] = usa_t_frame.date.apply(globe_date_convert)

# if USA Today missed the abstract, delete that row
usa_t_pattern_bad = 'Search | Saved Search | Login | Tips | FAQ | Pricing | My Account | Help | About | Terms'
usa_t_frame = usa_t_frame[~usa_t_frame['abstract'].str.contains(usa_t_pattern_bad)]

# clean wsj data
wsj_frame = pd.read_csv("./Data/Newspaper_data/WSJ_data/WSJ_obama_romney_newest.csv", encoding = "UTF-8")
wsj_frame['date'] = wsj_frame.date.apply(globe_date_convert)
wsj_frame.rename(columns={'snippet':'abstract'},inplace=True)
wsj_frame = wsj_frame[wsj_frame['date'] > datetime.date(2012,1,1)]
wsj_frame = wsj_frame[wsj_frame['date'] < datetime.date(2012,11,6)]

# combine dataframes
total_frame = globe_frame.append(guardian_frame,ignore_index=True)
total_frame = total_frame.append(chicago_frame,ignore_index=True)
total_frame = total_frame.append(usa_t_frame,ignore_index=True)
total_frame = total_frame.append(latimes_frame,ignore_index=True)
total_frame = total_frame.append(nydn_frame,ignore_index=True)
total_frame = total_frame.append(nyt_frame,ignore_index=True)
total_frame = total_frame.append(wp_frame,ignore_index=True)
total_frame = total_frame.append(newsday_frame,ignore_index=True)
total_frame = total_frame.append(pa_frame,ignore_index=True)
total_frame = total_frame.append(wsj_frame,ignore_index=True)

# clean the entire dataframe
total_frame = total_frame.drop(['Unnamed: 0', 'Unnamed: 0.1', 'keywords','subjects'],1)
total_frame['id'] = list(xrange(len(total_frame.index)))
total_frame['abstract'] = total_frame['abstract'].fillna('-')

def remove_na_wordcounts(wordcount):
    try: 
        wordcount = float(wordcount)
    except:
        wordcount = np.nan
        
    return wordcount

total_frame['word_count'] = total_frame['word_count'].apply(remove_na_wordcounts)

# split into Romney and Obama Subframes
def pres_filter(frame):
    total_frame = frame
    total_frame['candidate'] = ''
    romney_mask = ((total_frame.headline.str.contains("Romney")) & 
                   (~total_frame.headline.str.contains('Obama'))) | ((total_frame.abstract.str.contains("Romney")) 
                                                                     & (~total_frame.abstract.str.contains("Obama")) 
                                                                     & (~total_frame.headline.str.contains('Obama')))
    obama_mask = ((total_frame.headline.str.contains("Obama")) & 
                  (~total_frame.headline.str.contains('Romney'))) | ((total_frame.abstract.str.contains("Obama")) 
                                                                     & (~total_frame.abstract.str.contains("Romney")) 
                                                                     & (~total_frame.headline.str.contains('Romney')))
    total_frame.ix[romney_mask, 'candidate'] = "Romney"
    total_frame.ix[obama_mask, 'candidate'] = "Obama"
    return total_frame

total_frame = pres_filter(total_frame)
obama_frame = total_frame[total_frame['candidate'] == 'Obama']
romney_frame = total_frame[total_frame['candidate'] == 'Romney']
total_frame.to_csv("./Data/all_data.csv", encoding = "UTF-8")
romney_frame.to_csv("./Data/romney_data.csv", encoding = "UTF-8")
obama_frame.to_csv("./Data/obama_data.csv", encoding = "UTF-8")
```

```python
# plotting functions
def plot_histogram(data, percentiles = [5, 95], n_bins=100, frequency = True, legend = True,
                   lloc = 'upper right', datalabel = "", xlabel = "", ylabel = "", title = "",
                   xmin = 0, xmax = 100, alpha = 1, show = True, multiple = False):
    '''
    Generates a histogram of inputted data given number of bins and a list of percentiles to plot
    
    Input:
        data: a numpy array or pandas Series of data to plot
        precentiles: a list of vertical percentile bars to plot
        n_bins: number of bins
        frequency: make a frequency histogram. Otherwise, make a density histogram
        legend: show a legend T/F
        datalabel, xlabel, ylabel, title, etc.
    '''
    
    # plot a histogram of electoral vote relative frequencies
    if(frequency):
        hist = plt.hist(data, rwidth = None, bins=n_bins, label = datalabel, alpha = alpha)
    else:
        hist = plt.hist(data, weights=np.zeros_like(data) + 1./data.size, rwidth = None, bins=n_bins, 
                        alpha = alpha, label = datalabel)        
    
    # add vertical lines at victory threshold and result
    for idx, percentile in enumerate(percentiles):
        if(~multiple):
            plt.axvline(x = quantile(data, percentile), ymax = 0.5,  c = dark2_colors[idx + 1],
                    label = str(percentile) + 'th percentile')
    
    # add plot title and axis labels
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xlim(xmin, xmax)
    
    # add a legend, making the border width zero
    if (legend):
        legend = plt.legend(loc = lloc, shadow = False)
        legend.get_frame().set_linewidth(0)
    
    # remove borders
    remove_border();
    
    # show plot
    if(show):
        plt.show()

# declare a function to calculate values at percentile
def quantile(input_array, quantile):
    '''
    Returns the value of the kth quantile value in an unsorted Numpy array
    '''
    
    # sort input array
    input_array = np.sort(input_array)
    
    # find the index of the percentile value -- percentile times # elements-1
    index = (input_array.size-1)*float(quantile)/100
    
    # return the value closest to that index
    return input_array[round(index)]
```

##Sentiment Analysis
The data that we have collected so far does not have an intrinsic categorization built-in like the Rotten Tomatoes data from HW3. Instead, we will use three different precategorized datasets to evaluate each article on positivity, subjectivity, and support.

The positivity dataset is a set of 50,000 IMDB movie reviews from Stanford (url) that are presplit into testing and training data. While our dataset is composed of news articles, not movies, we will be using ngrams of at most length 3 and the general trend of whether a phrase connotes a positive feeling about a topic or not should be approximately the same.

The subjectivity dataset is composed of a set of IMDB movie summaries and Rotten Tomatoes reviews -- an IMDB summary is considered objective while Rotten Tomatoes data is considered subjective.

```python
# functions to load classifier training data

def load_positivity():
    '''
    Loads positivity data from Stanford IMDB dataset into memory as two lists: one text and the other
    binary 1 = positive, 0 = negative.
    
    This function concatenates the presorted testing and training data into one group.
    
    Output:
        quotes (text data), positivity (binary data)
    
    '''
    
    quotes, positivity = [], []
    
    for path in ['./Data/Training_Data/positivity/train/', './Data/Training_Data/positivity/test/']:
        # load all positive quotes
        for filename in os.listdir(path + 'pos/'):
            if(re.search(u'.txt', filename) is not None):
                with open(path + 'pos/' + filename) as f:
                    data="".join(line.rstrip() for line in f)
                quotes.append(plaintext(data))
                positivity.append(1)
        
        # load all negative quotes
        for filename in os.listdir(path + 'neg/'):
            if(re.search(u'.txt', filename) is not None):
                with open(path + 'neg/' + filename) as f:
                    data="".join(line.rstrip() for line in f)
                quotes.append(plaintext(data))
                positivity.append(0)
    
    return quotes, positivity

def load_support():
    '''
    Loads support / oppose data from political speeches dataset into memory as two lists: one text and the other
    binary 1 = support, 0 = no oppose.
    
    This function groups the presorted testing and training groups into one set.
    
    Output:
        quotes (text data), support (binary data)
    
    '''
    
    quotes, support = [], []
    
    for path in ['./Data/Training_Data/support_oppose/data_stage_one/training_set/', 
                 './Data/Training_Data/support_oppose/data_stage_one/test_set/']:
        # load all quotes
        for filename in os.listdir(path):
            if(re.search(u'.txt', filename) is not None):
                with open(path+filename) as f:
                    data="".join(line.rstrip() for line in f)
                quotes.append(plaintext(data))
                
                # determine if last letter is a Y or N for support binary data
                if filename.split('.txt')[0][-1] == 'Y':
                    support.append(1)
                else:
                    support.append(0)
    
    return quotes, support

def load_subjectivity():
    '''
    Loads subjectivity data from IMDB summary / Rotten Tomatoes review dataset into memory as two lists: one text and the other
    binary 1 = subjective, 0 = objective.
    
    Output:
        quotes (text data), support (binary data)
        
    There is no automatic split between testing and training data for this dataset.
    
    '''
    
    # load all subjective quotes
    subj = open('./Data/Training_Data/subjectivity/subj')
    subj_quotes, subjectivity = [], []
    
    for line in subj:
        subj_quotes.append(line.rstrip('\n'))
        subjectivity.append(1)
    subj.close()
    
    # load all objective quotes
    obj = open('./Data/Training_Data/subjectivity/obj')
    obj_quotes = []
    
    for line in obj:
        obj_quotes.append(line.rstrip('\n'))
        subjectivity.append(0)
    obj.close
    
    quotes = subj_quotes + obj_quotes
    
    # tranform subjective data into UTF-8
    for idx, item in enumerate(quotes):
        quotes[idx] = item.decode('latin1').encode('utf8')
    
    return quotes, subjectivity
```

####Vectorization
Having loaded in the training data, we now vectorize the data and convert it into a form that we can use for classification. There are two different vectorizers that we can use: CountVectorizer, which simply converts the string 'Romney is running for election' into a bag of words with a count of each word in the string. Tf-idf term weighting, however, re-weights the counts based on the frequency of words: e.g. words like 'the' or 'is' will receive a much lower weighting in the final analysis because these terms occur too frequently to be informative.

We will use sparse vectors throughout this analysis to save on memory space.

```python
# Fit a vectorizer to given data
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

def vectorize_data(quote_list, vectorizer = None, Tfidf = True, min_df = 1, 
                   ngram_range = (1,2), token_pattern = r'\b\w+\b'):
    '''
    Vectorizes given data using desired vectorizer object.
    
    Input:
        quote_list: list of data to vectorize
        vectorizer : CountVectorizer object (optional)
            A CountVectorizer object to use. If None,
            then create and fit a new CountVectorizer.
            Otherwise, re-fit the provided CountVectorizer
            using the provided data data
    
    Output:
        numpy array (dims: nreview, nwords)
            Bag-of-words representation for each quote.
    '''
    
    # if no vectorizer was passed, declare a vectorizer object
    if(vectorizer is None): 
        if(Tfidf == False):
            vectorizer = CountVectorizer(min_df = min_df, ngram_range = ngram_range, token_pattern = token_pattern)
        else:
            vectorizer = TfidfVectorizer(min_df = min_df, ngram_range = ngram_range, token_pattern = token_pattern)
    
    # build the vectorizer vocabulary
    vectorizer.fit(quote_list)

    # transform into bag of words
    X = vectorizer.transform(quote_list)
    
    return X.tocsc()
```

###Naive Bayes Training
We now use the vectorized data to fit a multinomial Naive Bayes classfier. First, however, we split the three datasets into testing and training data. We then fit the classifier and report the classification scores to determine the degree of overfitting.

In our preliminary analysis, we will use a CountVectorizer with unigrams. We will then look at more advanced options later.

```python
from sklearn.cross_validation import train_test_split
from sklearn.naive_bayes import MultinomialNB

def calibration_plot(clf, x, y, n_bins = 20, descriptor = ''):
    '''
    Plot the % correct of a classifier model vs. the actual % correct
    '''
    
    # generate the probability that a given classification is correct
    probability_correct_wrong = clf.predict_proba(x)
    
    # make an array of only the probability fresh data
    probability_correct = probability_correct_wrong[:,1]
    
    # place both lists into a DataFrame and sort them to faciliatate binning
    pred_results_df = pd.DataFrame({'prediction': probability_correct, 'actual': y}).sort('prediction', ascending = 1)
    
    # bin the data into n_bins bins, calculating the P(fresh) and %fresh
    bin_intervals = np.array([(1./n_bins) * i for i in range(n_bins + 1)])
    
    # suppress the warnings from reindexing
    warnings.filterwarnings("ignore")
    
    binned_counts = np.array([pred_results_df[pred_results_df.prediction > bin_intervals[i]]
                     [pred_results_df.prediction < bin_intervals[i + 1]].shape[0] for i in range(n_bins)])

    binned_results = np.array([float(pred_results_df[pred_results_df.actual == 1][pred_results_df.prediction > bin_intervals[i]]
                     [pred_results_df.prediction < bin_intervals[i + 1]].shape[0]) / binned_counts[i] for i in range(n_bins)])
    
    # calculate the uncertainty for later plotting
    fresh_sigma = np.sqrt((binned_results * (1-binned_results) / binned_counts))
    
    # plot the figure
    fig = plt.figure()
    
    # generate a plot of % true vs P(true)
    ax1 = fig.add_subplot(211)
    
    # add a gray y = x line
    x = np.linspace(0, 1, 100)
    ax1.plot(x, 100*x, '-', c='.75')
    
    # plot the data
    ax1.errorbar(bin_intervals[:n_bins], 100*binned_results, fmt = '.-', yerr = 100*fresh_sigma, ms = 12, c = dark2_colors[0])
    
    # set labels on axes & title
    plt.title('Percent ' + descriptor + ' vs. Prediction')
    plt.ylabel('Percent ' + descriptor)
    remove_border();
    
    # generate a plot of the count of 
    ax2 = fig.add_subplot(212, sharex=ax1)
    ax2.bar(bin_intervals[:n_bins], binned_counts, width = .04, color = dark2_colors[0])
    
    # set labels on axes & title
    plt.title('Count of Elements at Probability')
    plt.ylabel('Review Count')
    plt.xlabel('P(' + descriptor + ')')
    remove_border();
    
    # hide lower axis on top plot
    plt.setp(ax1.get_xticklabels(), visible=False)

    plt.show()

def compare_NB(Tfidf = False, ngram_range = (1,1), version = "CountVectorizer with unigrams and default values"):
    # vectorize the three datasets
    pos_X_vec = vectorize_data(pos_X, Tfidf = Tfidf, ngram_range = ngram_range)
    sup_X_vec = vectorize_data(sup_X, Tfidf = Tfidf, ngram_range = ngram_range)
    subj_X_vec = vectorize_data(subj_X, Tfidf = Tfidf, ngram_range = ngram_range)
    
    # split the datasets into training and testing
    test_size = 0.33
    pos_X_train, pos_X_test, pos_Y_train, pos_Y_test = train_test_split(pos_X_vec, pos_Y, test_size = test_size)
    sup_X_train, sup_X_test, sup_Y_train, sup_Y_test = train_test_split(sup_X_vec, sup_Y, test_size = test_size)
    subj_X_train, subj_X_test, subj_Y_train, subj_Y_test = train_test_split(subj_X_vec, subj_Y, test_size = test_size)
    
    # fit a MultinomialNB classifier to each dataset
    clf_pos, clf_sup, clf_subj = MultinomialNB(), MultinomialNB(), MultinomialNB()
    clf_pos.fit(pos_X_train, pos_Y_train)
    clf_sup.fit(sup_X_train, sup_Y_train)
    clf_subj.fit(subj_X_train, subj_Y_train)
    
    # Print the accuracy on the test and training datasets
    print "Scores of " + version + ":"
    print ""
    print "Positivity dataset:"
    print "Training set score: " + str(clf_pos.score(pos_X_train, pos_Y_train))
    print "Testing set score: " + str(clf_pos.score(pos_X_test, pos_Y_test))
    print ""
    print "Support dataset:"
    print "Training set score: " + str(clf_sup.score(sup_X_train, sup_Y_train))
    print "Testing set score: " + str(clf_sup.score(sup_X_test, sup_Y_test))
    print ""
    print "Subjectivity dataset:"
    print "Training set score: " + str(clf_subj.score(subj_X_train, subj_Y_train))
    print "Testing set score: " + str(clf_subj.score(subj_X_test, subj_Y_test))
    
    # plot calibration plots of fit
    calibration_plot(clf_pos, pos_X_test, pos_Y_test, n_bins = 20, descriptor = 'Positive')
    calibration_plot(clf_sup, sup_X_test, sup_Y_test, n_bins = 20, descriptor = 'Support')
    calibration_plot(clf_subj, subj_X_test, subj_Y_test, n_bins = 20, descriptor = 'Subjective')
```

To improve the fit of this classifier, we can use cross-validation to determine the best constants for fitting.

First, we will attempt to cross-validate by maximizing the testing set score. We will then attempt to maximize the log-liklihood and we will compare the results.

```python
def log_likelihood(clf, x_vals, y_vals):
    '''
    Computes the log liklihood of a dataset according to a Bayseian classifier
    '''
    
    # generate log liklihood scores
    logP_pos_neg = clf.predict_log_proba(x_vals)
    
    # separate logP(fresh) and logP(rotten) into two lists
    logP_neg = logP_pos_neg[:,0]
    logP_pos = logP_pos_neg[:,1]
    
    # place both lists into a DataFrame
    logP_scores_df = pd.DataFrame({'logP_neg': logP_neg, 'logP_pos': logP_pos, 'actual': y_vals})
    
    # calculate the sums of freshness log scores and rotten log scores
    sum_neg = logP_scores_df[logP_scores_df.actual == 0].logP_neg.sum()
    sum_pos = logP_scores_df[logP_scores_df.actual == 1].logP_pos.sum()
    
    return sum_pos + sum_neg
```
